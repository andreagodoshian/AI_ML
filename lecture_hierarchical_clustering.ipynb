```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
the_path = r"C:\Users\12489\Documents\Data\Mall_Customers.csv"

mall_data = pd.read_csv(the_path)
mall_data.head(3)
```

```python
slim_data = mall_data.iloc[:, [3,4]]
slim_data.head(3)
```

```python
import scipy.cluster.hierarchy as s_c_h

dendrogram = s_c_h.dendrogram(s_c_h.linkage(slim_data, method='ward'))

plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

agg_hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
agg_hc.fit_predict(slim_data)
```

```python
plt.scatter(slim_data['Annual Income (k$)'], slim_data['Spending Score (1-100)'], c=agg_hc.labels_)

plt.title("Clusters of Customers")
plt.xlabel('Annual Income (k$)')
plt.ylabel("Spending Score (1-100)")

plt.show()
```
